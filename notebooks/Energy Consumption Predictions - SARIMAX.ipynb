# Stationarity 

For this case, I chose to work with the daily series. Let's check and see if it's stationary or not using the adfuller test. 

```python
from statsmodels.tsa.stattools import adfuller
results = adfuller(daily_usage.usage)

# printing results
print('ADF Statistic:', results[0])
print('p-value:', results[1])
```

From the adfuller test that captures pretty well the "stationary" trend we note the following results:
- The ADF statistics are  quite negative which hints on stationarity 
- P-value is greater than 0.05 , therefore we cannot reject H0 stating non-stationarity. 

This series is thereof **non-stationary**. Before we start  to differentiate it to make it stationary and ready to forecast, we'll decompose the seasonal attributes first.

# Forecasting using ARIMA

## picking the number of differencing order

```python
results = adfuller(daily_usage)

# printing results
print('ADF Statistic:', results[0])
print('p-value:', results[1])
```

```python
results_ = adfuller(daily_usage.diff().dropna())

# printing results
print('ADF Statistic:', results_[0])
print('p-value:', results_[1])
```

Differecing our data one time seems to fully convert the series into a stationary one. We'll then use d=1 for our model.

## Detemining the orders of p and q 

```python
df_= daily_usage.diff().dropna()
fig, (ax1, ax2) = plt.subplots(2, 1, figsize =(8,8))
plot_acf(df_, lags=25, zero=False, ax=ax1)
plot_pacf(df_, lags=25, zero=False, ax=ax2);
```

By comparing the ACF and PACF plot, the first  lightly cuts off at lag 2 and the second tails off.This result doesn't allow us to choose the number or orders confidently. We'll use another approach depicting the good-fit of the model through AIC and BIC  information criteria

## AIC vs BIC

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
order_aic_bic=[]
for p in range(4):
    for q in range(4):
        try:
            model = SARIMAX(daily_usage, order=(p,1,q))
            results= model.fit(disp=0)
            print(p, q, results.aic, results.bic)
            order_aic_bic.append((p, q, results.aic, results.bic))
        except:
            print(p, q, None, None)
            order_aic_bic.append((p, q, None, None))

order_df = pd.DataFrame(order_aic_bic, columns =["p", "q", "aic", "bic"])
```

```python
#sorting by aic 
order_df.sort_values("aic")
```

```python
#sorting by aic 
order_df.sort_values("bic")
```

Luckily enough, AIC & BIC chose the same model of orders: p=1, q=1 and we have d=1 

## Fitting model

```python
model= SARIMAX(daily_usage, order=(1,1,1))
results = model.fit()
residuals= results.resid
```

## Model Diagnostics

### Residuals 

```python
residuals.mean()
```

### MAE

```python
mae = np.mean(np.abs(residuals))
mae
```

### diagnostics plot

```python
results.plot_diagnostics(figsize=(12,8));
```

```python
results.summary()
```

We note : 
- Standardized residuals plot : No obvious structures
- histogram : The histogram shows us the measured distribution; the orange line shows a smoothed version of this histogram; and the green line, shows a normal distribution. However, these 2 lines are not aligned. This needs rectification and another diagnostic
- The normal Q-Q plot: is another way to show how the distribution of the model residuals compares to a normal distribution. Our residuals are not quite normally distributed as the points at the very end don't lie along the red line
- The last plot is the correlogram, which is just an ACF plot of the residuals rather than the data. 95% of the correlations for lag greater than zero should not be significant. There is no significant correlation in the residuals, and this is OK. 

To confirm the normal distribution and autocorrelation assumptions of the residuals, we look at the summary : 
- prob(Q)= 0.96 = we cannot reject the null hypothesis that the residuals are uncorrelated. That's a good sign. 
- prob(JB)=0.00= We reject the null hypothesis that the residuals are normal. *We need to fix that*. 

## a step back in our Box Jekins

```python
daily_usage.isnull().sum()
```

```python
(daily_usage==np.inf).sum()
```

```python
# taking the sqrt of the data
daily_usage= np.sqrt(daily_usage)
```

```python
order_aic_bic=[]
for p in range(4):
    for q in range(4):
        try:
            model = SARIMAX(daily_usage, order=(p,1,q))
            results= model.fit(disp=0)
            print(p, q, results.aic, results.bic)
            order_aic_bic.append((p, q, results.aic, results.bic))
        except:
            print(p, q, None, None)
            order_aic_bic.append((p, q, None, None))

order_df = pd.DataFrame(order_aic_bic, columns =["p", "q", "aic", "bic"])
```

```python
#sorting by aic 
order_df.sort_values("aic")
```

```python
#sorting by aic 
order_df.sort_values("bic")
```

After transforming our data and taking its sqrt, both our info criteria tie down to the same model again, with the following number of orders; p=1, d=1, q=1

```python
model= SARIMAX(daily_usage, order=(1,1,1))
results = model.fit()
residuals= results.resid
```

```python
print('MAE :', np.mean(np.abs(residuals)))
print('Average Residuals :', residuals.mean())    
```

```python
results.plot_diagnostics(figsize=(12,8));
```

```python
results.summary()
```

According to the new model summary, the model meets the condition of independence of the residuals (no correlation) because the p-value of the Ljung-Box test (Prob(Q)) is greater than 0.05, so we cannot reject the null hypothesis of independence, but we cannot say for sure that the residual distribution is homoscedastic because the p-value of the Jarque-Bera test (Prob(JB)) is smaller than 0.05.


# Forecasting 

```python
#in sample prediction 
forecast=results.get_prediction(start= -100, dynamic=True)
#mean forecast
mean_forecast = forecast.predicted_mean
# conf intervals
conf_int= forecast.conf_int()
```

```python
# Selecting lower and upper confidence limits
lower_limits = conf_int.loc[:,'lower usage']
upper_limits = conf_int.loc[:,'upper usage']
```

```python
print(mean_forecast)
```

```python
# plotting the orginal daily usage data 
plt.plot(daily_usage, label='observed')

# plotting mean predictions
plt.plot(mean_forecast.index, mean_forecast, color='r', label='forecast')

plt.fill_between(lower_limits.index, lower_limits, 
               upper_limits, color='pink')

# set labels, legends and show plot
plt.xlabel('Date')
plt.ylabel('Energy Usage')
plt.legend()
plt.show()
```
